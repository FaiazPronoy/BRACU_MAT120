```python
import sympy as sp
import matplotlib.pyplot as  plt
import numpy as np
```

```python
f = lambda x: x**2 + 2*x + 1
f(2)
```

```python
x = sp.symbols("x")

f = x**2 + 2*x + 1

# print("f(x) =", f)
display(f)
```

```python
a, b, c = sp.symbols("a, b, c")

g = a**2/2 + 3*b - c/a

# print("g(a,b,c) =", g)
display(g)
```

```python
root = sp.sqrt(18)
display(root)
```

```python
#For_decimal_value
root = sp.sqrt(18)
print(root.evalf())
print(root.evalf(4)) #Significant number of decimal value
```

```python
eval = root.evalf()
print(sp.Rational(eval)) #Convert Rational

print(sp.Rational(eval).limit_denominator(100)) #For simplyfiying
```

```python
x = sp.symbols("x")

f = x**2 + 2*x + 1

display(f)

display(f.subs(x, "Value")) #Replace x with a variable or value

```

```python
F = sp.lambdify(x, f)
print(F(1))
```

```python
G = sp.lambdify((a, b, c), g) #G = sp.lambdify([a, b, c], g)
print(G(1,2,3))
```

```python
f = x**2 + 2*x + 1
print("f(x) =", f)

dom = np.linspace(-5, 10, 1000)
ran = sp.lambdify(x, f)


plt.plot(dom, ran(dom), label= "f(x)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.axhline(y= 0, c = "red")
plt.axvline(x=0, c= "red")

plt.legend()
plt.grid()
```

```python
f = x**3 + 4*x**2 + 5
print("f(x) =", f)

dom = np.arange(-5, 10, 0.00001)
ran = sp.lambdify(x, f)


plt.plot(dom, ran(dom), label= "f(x)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.axhline(y= 0, c = "red")
plt.axvline(x=0, c= "red")

plt.legend()
plt.grid()
```

```python
f1 = x**2 + 2*x + 1
f2 = x**3 + 4*x**2 + 5
print("f1(x) =", f1)
print("f2(x) =", f2)


dom1 = np.linspace(-5, 10, 1000)
dom2 = np.arange(-5, 10, 0.00001)
ran1 = sp.lambdify(x, f1)
ran2 = sp.lambdify(x, f2)


plt.plot(dom1, ran1(dom1), label= "f1(x)")
plt.plot(dom2, ran2(dom2), label= "f2(x)")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.axhline(y= 0, c = "red")
plt.axvline(x=0, c= "red")

plt.legend()
plt.grid()

# plt.ylim(-20, 50) For zooming
```
