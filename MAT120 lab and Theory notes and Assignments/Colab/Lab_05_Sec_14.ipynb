```python
import sympy as sp
import numpy as np

from scipy.integrate import quad
#or
#import scipy.integrate as sc

import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
```

```python
f = sp.exp(x)*sp.cos(x)
print('f(x) =')
display(f)

#Indefinite:
sol = sp.integrate(f, x)
display(sp.Integral(f,x))
print('=')
display(sol)+
print('=')
display(sol.simplify())

#Definite:
a = 0
b = 2*np.pi

sol2 = sp.integrate(f, (x, a, b))

display(sp.Integral(f,(x, a, 2*sp.pi)))
print('=')
display(sol2)
```

```python
#Quad

f = sp.exp(x)*sp.cos(x)
print('f(x) =')
display(f)

fl = sp.lambdify(x,f)

sol3, err = quad(fl, a, b)    #[solution, error]

print('Solution: ', sol3)
print('Error: ', err)
```

```python
def dydx(y,x):
  return -y**2/x**2

def EM(x0, xf, y0, N):

  dx = (xf - x0)/N

  sum = 0
  x_val = []
  y_val = []

  for i in range(0,N):
    x_val.append(x0)
    y_val.append(y0)

    y0 += dydx(y0,x0)*dx
    x0 += dx

  return x_val, y_val

x0 = 1
y0 = 1/2
xf = 10
N = 1000

xv, yv = EM(x0, xf, y0, N)

plt.plot(xv, yv)
```
