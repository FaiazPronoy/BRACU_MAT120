```python
import sympy as sp
import numpy as np
```

```python
x = sp.symbols('x')
```

```python
def trapezoidal_rule(f, a, b, N):

  h = (b-a)/N

  #Part 1:
  sum = 1/2 * (f.subs(x,a) + f.subs(x,b))

  #Part 2:
  x_i = a

  for i in range(1, N-1):
    x_i += h

    sum += f.subs(x,x_i)

  return sum*h

f = sp.exp(x)*sp.sin(x)
print('f(x) =')
display(f)

a = 0
b = 2*np.pi
N = 1000

sol = trapezoidal_rule(f,a,b,N)

display(sp.Integral(f,(x,a,sp.pi)))
print('=', sol)

```

```python
def MC(f, a, b, N):

  sum = 0

  r = (b-a)/N

  for i in range(0,N):
    x_i = np.random.uniform(a,b)

    sum += f.subs(x,x_i)

  return sum*r

f = sp.exp(x)*sp.sin(x)
print('f(x) =')
display(f)

a = 0
b = 2*np.pi
N = 10000

sol = MC(f,a,b,N)

display(sp.Integral(f,(x,a,sp.pi)))
print('=', sol)
```

```python
def MC2(f, a, b, N):

  y = sp.lambdify(x,f)
  x_v = np.linspace(a,b,N)

  y_max = y(x_v).max()
  cnt = 0

  for i in range(0,N):
    rx = np.random.uniform(a,b)
    ry = np.random.uniform(0, y_max)

    if ry <= y(rx):
      cnt += 1

  area_box = (b-a)*y_max
  result = cnt/N * area_box

  return result

f = sp.exp(x)*sp.sin(x)
print('f(x) =')
display(f)

a = 0
b = 2*np.pi
N = 10000

sol = MC2(f,a,b,N)

display(sp.Integral(f,(x,a,sp.pi)))
print('=', sol)
```
