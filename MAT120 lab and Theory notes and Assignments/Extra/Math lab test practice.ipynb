```python
import math
import matplotlib.pyplot as plt

def factorial(x):
    return math.gamma(x+1)

def trapezoidal_rule(expr, a, b, h):
    n = int((b - a) / h)
    sum = 0.5 * (expr(a) + expr(b))

    for i in range(1, n):
        x_i = a + i * h
        sum += expr(x_i)

    return sum * h

a = 0
b = 1
h = 0.01

x_values = []
y_values = []

for i in range(101):
  x_values.append(i*0.01)

for j in x_values:
  y_values.append(factorial(j))

plt.plot(x_values, y_values)
plt.title('Factorial function')
plt.xlabel('x')
plt.ylabel('x!')
plt.show()



integral = trapezoidal_rule(factorial, a, b, h)
print(f'Result: {round(integral, 4)}.')
```

```python
import numpy as np
import sympy as sp
from scipy.optimize import minimize

# Define the variable and function
x = sp.symbols('x')
f = 3 + 0.5 * sp.sin(np.pi * x / 5) * x * sp.exp(-x / 10)

# Lambdify the function
fl = sp.lambdify(x, f, "numpy")

# Define the range
a = -15
b = 15

# Calculate the derivative of the function
df = sp.diff(f, x)

# Lambdify the derivative
dfl = sp.lambdify(x, df, "numpy")

# Calculate the second derivative of the function
df2 = sp.diff(df, x)

# Lambdify the second derivative
df2l = sp.lambdify(x, df2, "numpy")

# Define the bisection method
def bisection(f, a, b, tol=1e-5):
    c = a
    while ((b-a) >= tol):
        c = (a+b)/2
        if (f(c) == 0.0):
            break
        if (f(c)*f(a) < 0):
            b = c
        else:
            a = c
    return c

# Find the critical points
c_p = []
temp = 0
x_i = np.linspace(a, b, 400)

for i in x_i:
    root = bisection(dfl, temp, i)
    if np.abs(root - temp) > 1e-5:
        temp = root
        c_p.append(temp)

# Find the maximum value
max_val = []

for p in c_p:
    if df2l(p) < 0:
        max_val.append(fl(p))

print('The maximum value is: ', max(max_val))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, y):
  return np.cos(t) + np.exp(-t) * y

def rk4(f, y0, t0, tf, dt):

  t = np.arange(t0, tf+dt, dt)
  y = np.zeros_like(t)
  y[0] = y0

  for i in range(1, len(t)):
    k1 = f(t[i-1], y[i-1])
    k2 = f(t[i-1] + dt/2, y[i-1] + k1*dt/2)
    k3 = f(t[i-1] + dt/2, y[i-1] + k2*dt/2)
    k4 = f(t[i-1] + dt, y[i-1] + k3*dt)
    y[i] = y[i-1] + dt/6 * (k1 + 2*k2 + 2*k3 + k4)
  return t, y

y0 = 0
t0 = 0
tf = 20
dt = 0.05

t, y = rk4(f, y0, t0, tf, dt)

# Plot the results
plt.plot(t, y)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Runge-Kutta 4th Order Approximation')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the system of differential equations
def predator_prey_system(y, t, a, b, c, d):
    X, Y = y
    dXdt = a * X + b * X * Y
    dYdt = c * X + d * X * Y
    return [dXdt, dYdt]

# Set the parameter values
a = 0.04
b = 0.0005
c = -0.1
d = 0.005

# Set the initial conditions
X0 = 50
Y0 = 10
initial_conditions = [X0, Y0]

# Set the time points at which the solution will be evaluated
t = np.linspace(0, 365, 1000)

# Solve the system of differential equations
solution = odeint(predator_prey_system, initial_conditions, t, args=(a, b, c, d))

# Extract the solutions for X and Y
X_solution = solution[:, 0]
Y_solution = solution[:, 1]

# Plot the solutions
plt.plot(t, X_solution, label='Prey (X)')
plt.plot(t, Y_solution, label='Predator (Y)')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
import numpy as np

def f(t, y):
    # Define prey and predator populations
    x, y = y

    # Define system of differential equations
    dxdt = a * x + b * x * y
    dydt = c * x + d * x * y

    return np.array([dxdt, dydt])

# Define parameters
a = 0.04
b = 0.0005
c = -0.1
d = 0.005

# Define initial conditions
y0 = np.array([50, 10]) # [X0, Y0]

def rk4(f, y0, t0, tf, dt):
    # Define time interval and step size
    t = np.arange(t0, tf+dt, dt)
    y = np.zeros((len(t), len(y0)))
    y[0] = y0

    # Implement Runge-Kutta 4th order method
    for i in range(1, len(t)):
        k1 = f(t[i-1], y[i-1])
        k2 = f(t[i-1] + dt/2, y[i-1] + k1*dt/2)
        k3 = f(t[i-1] + dt/2, y[i-1] + k2*dt/2)
        k4 = f(t[i-1] + dt, y[i-1] + k3*dt)
        y[i] = y[i-1] + dt/6 * (k1 + 2*k2 + 2*k3 + k4)
    return t, y.T

# Solve the system numerically
t, y = rk4(f, y0, 0, 365, 1)

# Extract prey and predator populations
x, y = y

import matplotlib.pyplot as plt

# Plot prey and predator populations
plt.plot(t, x, label='X (Prey)')
plt.plot(t, y, label='Y (Predator)')

# Add labels and grid
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the system of differential equations
def dydt(t, y, a, b, c, d):
    X, Y = y
    return np.array([a*X + b*X*Y, c*X + d*X*Y])

# Euler-Maruyama method
def EM(t0, tf, y0, N, a, b, c, d):
    dt = (tf - t0)/N
    t_values = np.linspace(t0, tf, N)
    X_values = []
    Y_values = []

    for t in t_values:
        X_values.append(y0[0])
        Y_values.append(y0[1])
        y0 += dt*dydt(t, y0, a, b, c, d)

    return t_values, X_values, Y_values

# Parameters
a = 0.04
b = 0.0005
c = -0.1
d = 0.005
y0 = [50, 10]  # Initial conditions
N = 365  # Number of steps
t0 = 0  # Start time
tf = 365  # End time

# Solve the system of differential equations
t_values, X_values, Y_values = EM(t0, tf, y0, N, a, b, c, d)

# Plot the results
plt.plot(t_values, X_values, label='Prey (X)')
plt.plot(t_values, Y_values, label='Predator (Y)')
plt.legend()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbols
t = sp.symbols('t')
X = sp.Function('X')(t)
Y = sp.Function('Y')(t)

# Parameters
a = 0.04
b = 0.0005
c = -0.1
d = 0.005

# System of differential equations
eq1 = sp.Eq(sp.diff(X, t), a*X + b*X*Y)
eq2 = sp.Eq(sp.diff(Y, t), c*X + d*X*Y)

# Solve the system of differential equations
sol = sp.dsolve((eq1, eq2))

# Extract the solutions for X and Y
X_sol = sol[0].rhs
Y_sol = sol[1].rhs

# Define initial conditions
X0 = 50
Y0 = 10

# Substitute initial conditions
X_particular = sp.lambdify(t, X_sol.subs({X.subs(t, 0): X0}))
Y_particular = sp.lambdify(t, Y_sol.subs({Y.subs(t, 0): Y0}))

# Define time interval and step size
t_min = 0
t_max = 365
dt = 0.1

# Numerical solution
def RK4(f, y0, t0, t_max, dt):
    t = np.arange(t0, t_max + dt, dt)
    y = np.zeros((len(t), len(y0)))
    y[0] = y0

    for i in range(1, len(t)):
        k1 = f(t[i - 1], y[i - 1])
        k2 = f(t[i - 1] + dt / 2, y[i - 1] + k1 * dt / 2)
        k3 = f(t[i - 1] + dt / 2, y[i - 1] + k2 * dt / 2)
        k4 = f(t[i - 1] + dt, y[i - 1] + k3 * dt)
        y[i] = y[i - 1] + dt / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
    return t, y

# Define the function for the system
def f(t, y):
    x, y = y
    return [a * x + b * x * y, c * x + d * x * y]

# Solve numerically with initial conditions
t, y = RK4(f, np.array([X0, Y0]), t_min, t_max, dt)

# Extract numerical solutions for X and Y
X_numerical = y[:, 0]
Y_numerical = y[:, 1]

# Plot the solutions
plt.plot(t, X_numerical, label='X (numerical)')
plt.plot(t, X_particular(t), label='X (analytical)')
plt.plot(t, Y_numerical, label='Y (numerical)')
plt.plot(t, Y_particular(t), label='Y (analytical)')
plt.legend()
plt.xlabel('Time (days)')
plt.ylabel('Population')
plt.title('Predator-Prey Model')
plt.grid(True)
plt.show()

print("General solution:")
print(X_sol)
print(Y_sol)
print("Particular solution:")
print(X_particular)
print(Y_particular)
```
