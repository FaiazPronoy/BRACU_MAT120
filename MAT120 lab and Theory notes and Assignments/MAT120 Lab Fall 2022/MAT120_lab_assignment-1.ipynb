```python
#y= x3−x2-x in the interval [-5, 6]
from sympy import*
import numpy as np
import matplotlib.pyplot as plt
x= symbols('x')
f=x**3-x**2-x
df= diff(f,x)
sorted=solve(df,x)
df2=diff(df,x)
if (len(sorted)>0):
  for i in range(len(sorted)):
    current=df2.subs(x,float(sorted[i]))
    if current<0:
      print('The maxima is:', sorted[i])
    elif current>0:
      print('The minima is:', sorted[i])
else:
  print('there is no maxima or minima')

f=lambdify(x,f)
var=np.arange(-5,7)
y=f(var)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('x**3-x**2-x')
plt.plot(var,y)
plt.show()
```
