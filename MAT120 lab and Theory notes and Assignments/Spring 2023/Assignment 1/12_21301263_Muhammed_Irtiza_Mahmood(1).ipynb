```python
#Task 1(a)
#Import sympy
from sympy import*
#Import numpy
import numpy as np
#Import matplotlib
import matplotlib.pyplot as plt
#Making or assuming a symbol
x= symbols('x')
#Writing a symbolic expression
f=x**4-8*x**3+22*x**2-24*x+5
#Taking the first derivative of the function
df= diff(f,x)
#Sorting and substituting
sorted=solve(df,x)
#Determination of minima and maxima
df2=diff(df,x)
# If-else condition and for loop
if (len(sorted)>0):
  for i in range(len(sorted)):
    current=df2.subs(x,float(sorted[i]))   #Print output
    if current<0:
      print('The maxima is:', sorted[i])   #Print output
    elif current>0:
      print('The minima is:', sorted[i])   #Print output
else:
  print('there is no maxima or minima')

f=lambdify(x,f)
```

```python
#task 3(a)
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
#Defines the function and its derivatives
x=symbols("x")
y=(x**3)-(3*x)+2
y1=diff(y,x)
y2=diff(y1,x)
y3=diff(y2,x)

#convert the sympy expressions into lambda functions
y_func=lambdify(x,y)
y1_func=lambdify(x,y1)
y2_func=lambdify(x,y2)
y3_func=lambdify(x,y3)

#Array of x values
x=np.arange(-3,3,1) 
#Evaluating function values for x
y_vals=y_func(x)
y1_vals=y1_func(x)
y2_vals=y2_func(x)

#Creating the plot
plt.plot(x,y_func(x),label="Function")
plt.plot(x,y1_func(x), label="First derivative")
plt.plot(x,y2_func(x),label="Seond derivative")
plt.plot(x,[y3]*len(x),label="Third derivative")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Graph")
plt.legend()
plt.grid()
plt.show()
```
