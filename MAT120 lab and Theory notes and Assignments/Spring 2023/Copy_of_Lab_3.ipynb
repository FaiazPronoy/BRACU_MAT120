##Plotting graphs

```python
#importing matplotlib
import matplotlib.pyplot as plt
  
# x & y axis values
x = (1,3,5,7,9)
y = (2,4,6,8,10)
  
# plotting the graph
plt.plot(x, y)
  
# naming axes
plt.xlabel('x - axis')
plt.ylabel('y - axis')
  
# title
plt.title('hello graph!')
  
# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,11,3)
y = 4*x**2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title("parabolic function")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10,11)
f = 2*x**3+ x**2 + 3*x+ 5 

#plotting
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f')
plt.title("Cubic function")
plt.show()
```

```python
#lets mix these two up

#importing library modules
import matplotlib.pyplot as plt
import numpy as np

#defining variables
x = np.arange(-10,11)
y = 4*x**2
f = 2*x**3+ x**2 + 3*x+ 5 

#plotting
plt.plot(x,y, label="parabola")
plt.plot(x,f, label="cubic", color="red")
plt.xlabel('x')
plt.ylabel('y,f')
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

###Plotting A Circle! 

```python
import matplotlib.pyplot as plt 
import numpy as np

r=int(input("Enter radius: "))
t=np.linspace(0, 2*np.pi)
x= r*np.cos(t)
y= r*np.sin(t)

plt.plot(x,y)
plt.axis('equal')
plt.show()
```

##Integration

```python
from sympy import *

x=symbols('x')
f=x**3

intf =integrate(f,x)
print(intf)

intf2=integrate(intf, x)
print(intf2)
```

```python
#lets mix these two up

#importing library modules
import matplotlib.pyplot as plt
import numpy as np

#defining variables
x = np.arange(-10,11)
y = 4*x**2
f = 2*x**3- x**2 + 3*x+ 5 

#plotting
plt.plot(x,y, label="parabola")
plt.plot(x,f, label="cubic", color="red")
plt.xlabel('x')
plt.ylabel('y,f')
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
# Ellipse plotting
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=1.     #x-position of the center
v=0.5    #y-position of the center
a=5     #radius on the x-axis
b=1.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')
plt.show()
```

```python
#Hyperbola plotting
import numpy as np
from matplotlib import pyplot as plt
x=np.linspace(-100,100)
a=1
b=2
y=np.sqrt((x/a)**2-1*b**2)
plt.plot(x,y)
plt.plot(x,-y)
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('graph of hyperbola')
plt.grid(True)
plt.show()
```
