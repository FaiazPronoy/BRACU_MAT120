The equation of a forced dampened harmonic oscillator is \\
\begin{equation} 
\frac{d^2x}{dt^2}+2\gamma \frac{dx}{dt}+k^2x=F(t)
\end{equation}
1. Plot x vs t. \\
2. Plot x vs t for different values of $\gamma$.

```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np

def dydt(y, t):
  return [y[1], (2*g*y[0]+k**2*A*np.sin(w*t))/(m-1)]

y0=[1,0]
g=-10 #damping constant
k=0.5 #spring constant
m=10 #mass of the oscillating particle
A= 5 #Amplitude
w=100 #angular frequency
ts=np.arange(0,10, 0.1)
yt=odeint(dydt, y0, ts)
ys=yt[:,0]
plt.plot(ts,ys)
plt.show()
```
