```python
#Task 1 (a)
#Import sympy
from sympy import*
#Import numpy
import numpy as np
#Import matplotlib
import matplotlib.pyplot as plt
#Making or assuming a symbol
x= symbols('x')
#Writing a symbolic expression
f=x**4-8*x**3+22*x**2-24*x+5
#Taking the first derivative of the function
df= diff(f,x)
#Sorting and substituting
sorted=solve(df,x)
#Determination of minima and maxima
df2=diff(df,x)
# If-else condition and for loop
if (len(sorted)>0):
  for i in range(len(sorted)):
    current=df2.subs(x,float(sorted[i]))   
    if current<0:
      print('The maxima is:', sorted[i])   
    elif current>0:
      print('The minima is:', sorted[i])   
else:
  print('there is no maxima or minima')

f=lambdify(x,f)
```

```python
#Task 3(a)
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
x=symbols("x")
f=(x**3)-(3*x)+2
#Differentation
f1=diff(f,x)
f2=diff(f,x,2)
f3=diff(f,x,3)

#converting the sympy expressions into lambda functions
f_func=lambdify(x,f)
f1_func=lambdify(x,f1)
f2_func=lambdify(x,f2)
f3_func=lambdify(x,f3)

#Array of x values
x=np.arange(-4,4,1) 

#Evaluating function values for x
f_value=f_func(x)
f1_value=f1_func(x)
f2_value=f2_func(x)

#Plotting on graph
plt.plot(x,f_func(x),label="Main function")
plt.plot(x,f1_func(x), label="1st derivative",color="teal")
plt.plot(x,f2_func(x),label="2nd derivative",color="purple")
plt.plot(x,[f3]*len(x),label="3rd derivative")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plot")
plt.legend()
plt.show()
```

```python
#Task 4
from sympy import *
import math
import matplotlib.pyplot as plt
import numpy as np

s=symbols("s")
s=np.arange(0,2*math.pi,math.pi/180)

#Circle part
a=np.cos(s)
b=np.sin(s)

#Ellipses Part
x=4*np.cos(s)
y=2*np.sin(s)

#Plot
plt.plot(x+y,y)
plt.plot(-(y+x),y)
plt.plot(y,x)
plt.plot(a,b)

plt.axis("equal")
plt.title("Basic Atom Shape")
plt.show()
```
