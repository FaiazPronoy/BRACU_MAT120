##Area under a curve

```python
#defining function
def f(t):
  return (4*t**2)

n=100 #interval
x1=1 #lower limit
x2=2 #upper limit

dx=(x2-x1)/n 

#starting points
A=0
x=x1

while x<=x2:
  dA=f(x)*dx
  A=A+dA
  x=x+dx

print("Area:",A) 
```

```python
#plotting 
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-5,5,0.1)
y=4*x**2

plt.plot(x,y) 
plt.fill_between(x,y, where=[(x>-5) and (x<5) for x in x])
plt.show()
```

##Area between two curves

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.optimize import fsolve,root
```

```python
x=np.linspace(-5,5,100)
y=lambda x: x+5
z=lambda x: x**2
h=lambda x: y(x)-z(x)

plt.plot(x,y(x),x,z(x))
plt.show()
```

```python
roots=root(h,(-5,5)).x
x1=roots[0]
x2=roots[1]
area=quad(h,x1,x2)
print(roots)
print("Area=", area)
```

```python
left=x>x1
right=x<x2
total=left & right

plt.plot(x,y(x),x,z(x))
plt.fill_between(x,y(x),z(x), where=total)
plt.show()
```
